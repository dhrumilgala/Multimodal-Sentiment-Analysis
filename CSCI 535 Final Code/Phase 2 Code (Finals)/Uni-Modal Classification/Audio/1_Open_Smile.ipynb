# **Feature extraction Using Open Smile**

```python
import os
import numpy as np
import soundfile as sf
import torch
import json
import pandas as pd
import opensmile

# Load the metadata
metadata_path = '/content/drive/MyDrive/2 Multi Modal/535/IEMOCAP_full_release/meta_data/meta_data.json'

with open(metadata_path, 'r') as f:
    metadata = json.load(f)

# Define a list to store the features and labels
features = []
labels = []
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv01b,
    feature_level=opensmile.FeatureLevel.Functionals,
    )

# Create an empty DataFrame to store the results
result_df = pd.DataFrame()
lable = []
```

```python
audio_dir = '/content/drive/MyDrive/2 Multi Modal/535/IEMOCAP_full_release'

# Loop through each item in the metadata file
for item in metadata['meta_data']:
    # Load the audio file and extract the relevant segment
    audio_file_path = os.path.join(audio_dir, item['path'])
    lable.append(metadata['labels'][item['label']])

    # Process the file with opensmile and store the result in a dictionary
    result_dict = smile.process_file(audio_file_path)
    # Convert the dictionary to a DataFrame
    result_df_file = pd.DataFrame(result_dict)
    # Append the file-specific DataFrame to the overall result DataFrame
    result_df = result_df.append(result_df_file)

result_df['emotion_label'] = lable
# Reset the index of the result DataFrame
result_df.reset_index(drop=True, inplace=True)
result_df.to_csv('/content/drive/MyDrive/2 Multi Modal/535/Open Smile/Final_Results_df.csv', index=False)
```

```python
# Create a new DataFrame to store the calculated values
calculated_values_df = pd.DataFrame()

# Iterate through the rows of the result_df DataFrame
for index, row in result_df.iterrows():
    # Extract the calculated values for each feature
    lableData = row["emotion_label"]
    spectral_flux_mean = row["spectralFlux_sma3_amean"]
    spectral_flux_stddev = row["spectralFlux_sma3_stddevNorm"]
    pitch_stddev = row["F0semitoneFrom27.5Hz_sma3nz_stddevNorm"]
    pitch_percentile20 = row["F0semitoneFrom27.5Hz_sma3nz_percentile20.0"]
    pitch_percentile80 = row["F0semitoneFrom27.5Hz_sma3nz_percentile80.0"]
    f0_semitone_mean = row['F0semitoneFrom27.5Hz_sma3nz_amean']
    f0_semitone_std = row['F0semitoneFrom27.5Hz_sma3nz_stddevNorm']
    loudness_mean = row['loudness_sma3_amean']
    loudness_std = row['loudness_sma3_stddevNorm']
    mfcc1_mean = row['mfcc1_sma3_amean']

    spectral_flux_mean = spectral_flux_mean.mean()
    spectral_flux_stddev = spectral_flux_stddev.mean()

    pitch_stddev = pitch_stddev.mean()
    pitch_range = int(pitch_percentile80 - pitch_percentile20)

    timber_pitch = f0_semitone_mean - mfcc1_mean

    loudness_range = loudness_std - loudness_mean

    mfcc_mean = result_df.filter(like='mfcc').mean(axis=1)
    mfcc_std = result_df.filter(like='mfcc').std(axis=1)
    
    calculated_values_dict = {
        'lable': lableData,
        'spectral_flux_mean': spectral_flux_mean,
        'spectral_flux_stddev': spectral_flux_stddev,
        'pitch_stddev': pitch_stddev,
        'pitch_percentile20': pitch_percentile20,
        'pitch_percentile80': pitch_percentile80,
        'f0_semitone_mean': f0_semitone_mean,
        'f0_semitone_std': f0_semitone_std,
        'loudness_mean': loudness_mean,
        'loudness_std': loudness_std,
        'mfcc1_mean': mfcc1_mean,
        'timber_pitch': timber_pitch,
        'loudness_range': loudness_range
    }
    
    # Append the dictionary as a row to the calculated_values_df DataFrame
    calculated_values_df = calculated_values_df.append(calculated_values_dict, ignore_index=True)

# Store the calculated values to a CSV file
calculated_values_df.to_csv('/content/drive/MyDrive/2 Multi Modal/535/Open Smile/Calculated_values_Open_Smile.csv', index=False)
```
